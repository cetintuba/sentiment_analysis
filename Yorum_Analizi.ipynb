Bu proje resturant yorumlarında olumlu olumsuz yorum duygu analizini yapar amaç  resturandın yorum veritabanından faydalanarak derecelendirme altyapısı olamayan bir sistem için olumlu olumsuz yorum üreten sistem tasarımıdır.

```python
# Veri analizi ve matematiksel işlemler
import pandas as pd
import numpy as np

# Metin ön işleme ve makine öğrenmesi
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Dil işleme
import re
```

```python
df = pd.read_csv('/content/restaurant.csv')
df.head()
```

```python
# Küçük harfe çevirme ve özel karakterleri kaldırma
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
# DataFrame içerisindeki ‘text’ sütunundaki her satıra clean_text fonksiyonunu uygular.
df['text'] = df['text'].apply(clean_text)
```

```python
# 'stars' sütunundan 'sentiment' sütunu oluşturma
# 3 ve üzeri yıldızları 'positive', altını 'negative' olarak etiketleme
df['sentiment'] = df['stars'].apply(lambda x: 'positive' if x >= 3 else 'negative')
x = df['text']      # Özellikler
y = df['sentiment']   # Hedef değişken

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
vectorizer = TfidfVectorizer()  # TF-IDF yöntemi: Kelimeleri sayılara çevirir. Kelimelerin önemini hesaplar
x_train_vec = vectorizer.fit_transform(x_train)
X_test_vec = vectorizer.transform(x_test)
```

```python
model = MultinomialNB()
model.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test_vec)

# Başarı ölçümü
accuracy = accuracy_score(y_test, y_pred)
print("Doğruluk (Accuracy):", accuracy)
print(classification_report(y_test, y_pred))
```

```python
new_review = "Yemekler çok lezzetli ve servis hızlıydı"
new_review_vec = vectorizer.transform([clean_text(new_review)])
prediction = model.predict(new_review_vec)
print("Tahmin:", prediction[0])
```
